### Q1. How many are there in each team and the percentage splitting with respect to the total employees

```python
team_count=data["Team"].value_counts()
team_count
```

```python
percentage=team_count/len(data)*100
percentage
```

```python
plt.bar(percentage.index,percentage.values,color='green')
plt.xlabel('Team Name')
plt.ylabel('percentage of employees')
plt.xticks(rotation = 90)
plt.title('Percentage of each Team')
plt.show()
```
